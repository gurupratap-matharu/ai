```python
import os
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa_filepath = "datasets/fifa.csv"
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=fifa_data)
```

<font color="943E58">We note that Brazil is the only country which stays at the number 1 position for 5 consecutive years!</font>

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=fifa_data['ARG'], label='Argentina')
sns.lineplot(data=fifa_data['FRA'], label='France')
plt.xlabel('Date')
```

```python
plt.figure(figsize=(16, 6))
plt.title('Argentina ranking over last decade')
sns.lineplot(data=fifa_data['ESP'], label='Spain')
plt.xlabel('Date')
```

```python
plt.figure(figsize=(16, 6))
sns.boxplot(data=fifa_data)
plt.title('Box and whisker plot to understand distribution in IQR')
plt.xlabel('Country')
plt.ylabel('Rank')
```

```python
plt.figure(figsize=(16, 6))
sns.boxenplot(data=fifa_data)
plt.xlabel('Country')
plt.ylabel('Rank')
plt.title('Boxen plot representing each country')
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(data=fifa_data)
```

```python
plt.figure(figsize=(16, 6))
sns.catplot(data=fifa_data, aspect=3)
```

```python
plt.figure(figsize=(16, 6))
sns.swarmplot(data=fifa_data)
```

```python
plt.figure(figsize=(16, 6))
sns.swarmplot(x='ARG', y='BRA', hue='FRA', data=fifa_data)
```
